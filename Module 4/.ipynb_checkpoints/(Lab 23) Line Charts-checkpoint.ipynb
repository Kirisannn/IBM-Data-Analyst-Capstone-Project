#### Step 1: Download the dataset


```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.ticker as mticker
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
import requests
from tqdm.notebook import tqdm
from urllib.parse import urlparse
import sqlite3

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

file_name = urlparse(file_path).path.strip('/').split('/')[-1]
```

#### Step 2: Import necessary libraries and load the dataset


#### Load the data


```python
# Load the Stack Overflow survey dataset
df = pd.read_csv(file_name)
```

#### Display the first few rows to understand the structure of the data


```python
# Display the first few rows of the dataset
df.head(3)
```

### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
df_1 = df.copy().loc[:,["ConvertedCompYearly", "Age"]].dropna()

age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older",
    "Prefer not to say"
]

df_1["Age"] = pd.Categorical(
    df_1["Age"],
    categories=age_order,
    ordered=True
)

df_1_median = df_1.groupby(by="Age", as_index=False, observed=False)["ConvertedCompYearly"].median()

sns.set_style(style="whitegrid")
plt.figure(figsize=(16,6))
sns.lineplot(
    x=df_1_median["Age"],
    y=df_1_median["ConvertedCompYearly"],
)
plt.title("Median Compensation by Age Group")
plt.ylabel("Converted Yearly Compensation")
plt.tight_layout()

plt.savefig("Lab 23 Plots/Task 1-1 Median Compensation by Age Group", bbox_inches='tight')
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
df_2 = df.copy().loc[(df["Age"] == "25-34 years old") | (df["Age"] == "35-44 years old") ,["ConvertedCompYearly", "Age"]].dropna()
df_2_median = df_2.groupby(by="Age", as_index=False, observed=False)["ConvertedCompYearly"].median()

sns.set_style(style="whitegrid")
plt.figure(figsize=(16,6))
sns.lineplot(
    x=df_2_median["Age"],
    y=df_2_median["ConvertedCompYearly"],
)
plt.title("Median Compensation by Age (25-44 years old)")
plt.ylabel("Converted Yearly Compensation")
plt.tight_layout()

plt.savefig("Lab 23 Plots/Task 1-2 Median Compensation by Age (25-44 years old)", bbox_inches='tight')
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
df_3 = df.copy().loc[:,["JobSatPoints_6", "WorkExp"]].dropna()
df_3["WorkExp"] = pd.to_numeric(df_3["WorkExp"], errors="coerce")

def experience_level(years):
    if pd.isna(years):
        return "Unknown"

    if years < 2:
        return "Beginner"
    elif 2 <= years < 5:
        return "Novice"
    elif 5 <= years < 10:
        return "Intermediate"
    elif 10 <= years < 20:
        return "Advanced"
    elif years >= 20:
        return "Expert"
    else:
        return "Unknown"

df_3["Experience"] = df_3["WorkExp"].apply(experience_level)

experience_order = [
    "Beginner",
    "Novice",
    "Intermediate",
    "Advanced",
    "Expert",
]

df_3["Experience"] = pd.Categorical(
    df_3["Experience"],
    categories=experience_order,
    ordered=True
)

sns.set_style(style="whitegrid")
plt.figure(figsize=(16,6))
sns.lineplot(
    x=df_3["Experience"],
    y=df_3["JobSatPoints_6"],
)
plt.title("Job Satisfaction by Experience Level")
plt.ylabel("Job Satisfaction")
plt.tight_layout()

plt.savefig("Lab 23 Plots/Task 2-1 Job Satisfaction by Experience Level", bbox_inches='tight')
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
df_4 = df.copy().loc[:,["ConvertedCompYearly","WorkExp"]].dropna()
df_4["WorkExp"] = pd.to_numeric(df_4["WorkExp"], errors="coerce")

sns.set_style(style="whitegrid")
plt.figure(figsize=(16,6))
sns.lineplot(
    x=df_4["WorkExp"],
    y=df_4["ConvertedCompYearly"],
)
plt.title("Converted Yearly Compensation by Experience")
plt.ylabel("Converted Yearly Compensation")
plt.tight_layout()

plt.savefig("Lab 23 Plots/Task 3-1 Converted Yearly Compensation by Experience", bbox_inches='tight')
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
df_5 = df.copy().loc[:,["JobSatPoints_6","WorkExp"]].dropna()
df_5["WorkExp"] = pd.to_numeric(df_5["WorkExp"], errors="coerce")

sns.set_style(style="whitegrid")
plt.figure(figsize=(16,6))
sns.lineplot(
    x=df_5["WorkExp"],
    y=df_5["JobSatPoints_6"],
)
plt.title("Job Satisfaction by Experience Level")
plt.ylabel("Job Satisfaction")
plt.tight_layout()

plt.savefig("Lab 23 Plots/Task 3-2 Job Satisfaction by Experience Level", bbox_inches='tight')
plt.show()
```
