## Setup: Working with the Database
**Install the needed libraries**


```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.ticker as mticker
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
import requests
from tqdm.notebook import tqdm
from urllib.parse import urlparse
import sqlite3

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:,.2f}'.format)
```

**Download and connect to the database file containing survey data.**


To start, download and load the dataset into a `pandas` DataFrame.



```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

file_name = urlparse(file_path).path.strip('/').split('/')[-1]
```

```python
# Load the Stack Overflow survey dataset
df = pd.read_csv(file_name)

# Display the first few rows of the dataset
df.head(3)
```

### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
df_1 = df.copy().loc[:,["ConvertedCompYearly"]].dropna()

plt.figure(figsize=(16,6))
sns.histplot(x=df_1.loc[df_1["ConvertedCompYearly"] <= df_1["ConvertedCompYearly"].quantile(0.99),"ConvertedCompYearly"], bins=100, kde=True)
# plt.xscale("log")
plt.title("Distribution of Converted Yearly Compensation (Excluding Top 1%)")
plt.xlabel("ConvertedCompYearly")
plt.ylabel("Frequency")
plt.savefig("Lab 24 Plots/Step 1-1 Distribution of Converted Yearly Compensation (Excluding Top 1%)")
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
df_2 = df.copy().loc[:,["Age"]].dropna()

def age_group_midpoint(age_str):
    if pd.isna(age_str):
        return None

    mapping = {
        "Under 18 years old": 15,          
        "18-24 years old": int((18 + 24) / 2),
        "25-34 years old": int((25 + 34) / 2),
        "35-44 years old": int((35 + 44) / 2),
        "45-54 years old": int((45 + 54) / 2),
        "55-64 years old": int((55 + 64) / 2),
        "65 years or older": int(65),
        "Prefer not to say": None
    }

    return mapping.get(age_str, None)

df_2["AgeMidpoint"] = df_2["Age"].apply(age_group_midpoint)

plt.figure(figsize=(16,4))
sns.boxplot(x=df_2["AgeMidpoint"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.savefig("Lab 24 Plots/Step 1-2 Age Distribution")
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
df_3 = df.copy().loc[:,["Age", "ConvertedCompYearly"]].dropna()
df_3["AgeMidpoint"] = df_3["Age"].apply(age_group_midpoint)

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
sns.scatterplot(x=df_3["AgeMidpoint"], y=df_3["ConvertedCompYearly"])
plt.title("Compensation by Age")
plt.xlabel("Age")
plt.ylabel("Compensation")
plt.savefig("Lab 24 Plots/Step 2-1 Compensation by Age")
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
df_4 = df.copy().loc[:,["ConvertedCompYearly", "JobSatPoints_6", "Age"]].dropna()
df_4["AgeMidpoint"] = df_4["Age"].apply(age_group_midpoint) * 100

plt.figure(figsize=(14,6))
sns.scatterplot(
    x=df_4["JobSatPoints_6"],
    y=df_4["ConvertedCompYearly"],
    size=df_4["AgeMidpoint"],
    sizes=(50, 500),
    alpha=0.4,
    legend=False,
    color='teal',
    edgecolors='black',
)

plt.title("Compensation by Job Satisfaction and Age")

plt.savefig("Lab 24 Plots/Step 2-2 Compensation by Job Satisfaction and Age")
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
df_5 = df.copy().loc[:,["MainBranch"]].dropna()

df_5_counts = df_5["MainBranch"].value_counts(dropna=False)

plt.figure(figsize=(8,8))
sns.barplot(
    x=df_5_counts.values,
    y=df_5_counts.index
)
plt.title("Distribution of Primary Roles")
plt.xlabel("Count")
plt.ylabel("Primary Role")

plt.savefig("Lab 24 Plots/Step 3-1 Distribution of Primary Roles")
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
df_6 = df.copy().loc[:,["LanguageWantToWorkWith"]].dropna()
df_6["LanguageWantToWorkWith"] = df_6["LanguageWantToWorkWith"].str.split(";")
df_6 = df_6.explode("LanguageWantToWorkWith")
df_6_counts = df_6["LanguageWantToWorkWith"].value_counts().head(5)

plt.figure(figsize=(8,5))
sns.barplot(
    x=df_6_counts.index,
    y=df_6_counts.values
)
plt.title("Top 5 Most Desired Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Count")

plt.savefig("Lab 24 Plots/Step 3-2 Top 5 Most Desired Programming Languages")
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
df_7 = df.copy().loc[:,["JobSatPoints_6", "JobSatPoints_7", "Age"]].dropna()

age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older",
    "Prefer not to say"
]

df_7["Age"] = pd.Categorical(
    df_7["Age"],
    categories=age_order,
    ordered=True
)

df_7_grouped = df_7.groupby(by="Age", as_index=False, observed=False)[["JobSatPoints_6", "JobSatPoints_7"]].median()

plt.figure(figsize=(16, 6))
plt.bar(
    x=df_7_grouped["Age"],
    height=df_7_grouped["JobSatPoints_6"],
    bottom=None,
    label="JobSatPoints_6",
    color="tab:blue",
    width=0.8,
    align="center"
)

plt.bar(
    x=df_7_grouped["Age"],
    height=df_7_grouped["JobSatPoints_7"],
    bottom=df_7_grouped["JobSatPoints_6"],
    label="JobSatPoints_7",
    color="tab:orange",
    width=0.8,
    align="center"
)

plt.xlabel(xlabel="Age Group")
plt.ylabel(ylabel="Median Job Satisfaction Points")
plt.title(label="Median Job Satisfaction Composition by Age Group")
plt.legend(loc='upper right', title='Satisfaction Metric')
plt.tight_layout()

plt.savefig("Lab 24 Plots/Task 3-3 Median Job Satisfaction Composition by Age Group")
plt.show()
```

```python
df_7 = df.copy().loc[:,["JobSatPoints_6", "JobSatPoints_7", "Age"]].dropna()

age_order = [
    "Under 18 years old",
    "18-24 years old",
    "25-34 years old",
    "35-44 years old",
    "45-54 years old",
    "55-64 years old",
    "65 years or older",
    "Prefer not to say"
]

df_7["Age"] = pd.Categorical(
    df_7["Age"],
    categories=age_order,
    ordered=True
)

df_7_grouped = df_7.groupby(by="Age", as_index=False, observed=False)[["JobSatPoints_6", "JobSatPoints_7"]].median()

fig, ax = plt.subplots(figsize=(16, 6))
bars_6 = ax.bar(
    x=df_7_grouped["Age"],
    height=df_7_grouped["JobSatPoints_6"],
    label="JobSatPoints_6",
    color="tab:blue",
    width=0.8,
    align="center"
)

bars_7 = ax.bar(
    x=df_7_grouped["Age"],
    height=df_7_grouped["JobSatPoints_7"],
    bottom=df_7_grouped["JobSatPoints_6"],
    label="JobSatPoints_7",
    color="tab:orange",
    width=0.8,
    align="center"
)

for container in ax.containers:
    ax.bar_label(container, label_type="center")


plt.xlabel(xlabel="Age Group")
plt.ylabel(ylabel="Median Job Satisfaction Points")
plt.title(label="Median Job Satisfaction Composition by Age Group")
plt.legend(loc='upper right', title='Satisfaction Metric')
plt.tight_layout()

plt.savefig("Lab 24 Plots/Task 3-3 Median Job Satisfaction Composition by Age Group")
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
df_8 = df.copy().loc[:, ["DatabaseHaveWorkedWith"]].dropna()
df_8["DatabaseHaveWorkedWith"] = df_8["DatabaseHaveWorkedWith"].str.split(";")
df_8 = df_8.explode("DatabaseHaveWorkedWith")

df_8_counts = df_8["DatabaseHaveWorkedWith"].value_counts().head(10)

sns.set_style("whitegrid")
plt.figure(figsize=(16,6))
ax = sns.barplot(
    x=df_8_counts.index,
    y=df_8_counts.values,
    hue=df_8_counts.index,
)

for container in ax.containers:
    ax.bar_label(container)

ax.set_title("Top 10 Databases Respondents Have Worked With")
ax.set_ylabel("Count")

plt.savefig("Lab 24 Plots/Task 3-4 Top 10 Databases Respondents Have Worked With")
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
df_9 = df.copy().loc[:,["ConvertedCompYearly", "Age"]].dropna()
df_9["Age"] = pd.Categorical(
    df_9["Age"],
    categories=age_order,
    ordered=True
)

df_9_grouped = df_9.groupby(by="Age", as_index=False, observed=False)["ConvertedCompYearly"].median()

sns.set_style("whitegrid")
plt.figure(figsize=(16,6))
ax = sns.barplot(
    x=df_9_grouped["Age"],
    y=df_9_grouped["ConvertedCompYearly"],
    hue=df_9_grouped["Age"],
)

for container in ax.containers:
    ax.bar_label(container)

ax.set_title("Median Compensation across Age Groups")

plt.savefig("Lab 24 Plots/Task 4-1 Median Compensation across Age Groups")
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
df_10 = df.copy().loc[:, ["Country"]].dropna()
df_10_count = df_10["Country"].value_counts().head(20)

sns.set_style("whitegrid")
plt.figure(figsize=(16,8))
ax = sns.barplot(
    x=df_10_count.index,
    y=df_10_count.values,
    hue=df_10_count.index,
    palette="tab20",
    legend=True
)

for container in ax.containers:
    ax.bar_label(container)

ax.set_title("Top 20 Countries of Respondents")
ax.set_ylabel("Count")
plt.xticks([])

plt.savefig("Lab 24 Plots/Task 4-2 Top 20 Countries of Respondents")
plt.show()
```
