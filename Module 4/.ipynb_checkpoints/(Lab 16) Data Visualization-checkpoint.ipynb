### Install and Import Necessary Python Libraries

Ensure that you have the required libraries installed to work with SQLite and Pandas:

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.ticker as mticker
import seaborn as sns
import requests
from tqdm.notebook import tqdm
from urllib.parse import urlparse

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
```

### Download the database file.


```python
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
data_name = urlparse(data_url).path.strip('/').split('/')[-1]
```

### Read the CSV File into a Pandas DataFrame

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Load the Stack Overflow survey dataset
df = pd.read_csv(data_name)

# Display the first few rows of the dataset
df.head(3)
```

### Create a SQLite Database and Insert the Data

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql(name='main', con=conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```

### Verify the Data in the SQLite Database
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
df_check.head()
```

Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
query = """
SELECT CompTotal FROM main
"""

compTotal = pd.read_sql(con=conn, sql=query)

# Limiting to 90th percentile since data being squashed into 1 bin near the left
compTotal = compTotal[compTotal["CompTotal"] <= compTotal["CompTotal"].quantile(0.90)]

plt.figure(figsize=(16,4))
sns.histplot(data=compTotal, x="CompTotal", bins=100, kde=True)
# plt.xscale("log")
plt.title("Distribution of Total Compensation (< 90th Percentile)")
plt.xlabel("CompTotal")
plt.ylabel("Frequency")
plt.savefig("Lab 16 - Histogram 90th percentile")
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
# Query data
query = """
SELECT Age from main
"""
age = pd.read_sql(con=conn, sql=query)
age.value_counts()
```

```python
# Query data
query = """
SELECT Age from main
"""
age = pd.read_sql(con=conn, sql=query)

# Clean
age['Converted_Age'] = age.copy()['Age']
def map_age(age_str):
    if pd.isna(age_str):
        return np.nan

    if age_str == "Under 18 years old":
        return 17
    elif age_str == "65 years or older":
        return 65

    # Extract all numbers from the string
    nums = list(map(int, re.findall(r'\d+', age_str)))

    if len(nums) == 2:
        # Regular range: take midpoint
        return (nums[0] + nums[1]) / 2
    else:
        return np.nan

age.loc[:,'Converted_Age'] = age.loc[:,'Converted_Age'].apply(map_age)
age = age.dropna(subset=["Converted_Age"])
age["Converted_Age"] = pd.to_numeric(age["Converted_Age"], errors="coerce")  # Non‑numeric → NaN
print(age['Converted_Age'].value_counts(dropna=False))
```

```python
plt.figure(figsize=(16,4))
sns.boxplot(
    data=age,
    x="Converted_Age"
)
plt.xlim(15, 70)
plt.xlabel("Age")
plt.xticks(np.arange(17, 66, 2))
plt.title("Age Distribution")
plt.savefig("Lab 16 - Boxplot (Age)")
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
query = """
SELECT Age, WorkExp FROM main
"""
age_exp = pd.read_sql(sql=query, con=conn)
age_exp.loc[:,"Age"] = age_exp.loc[:,"Age"].apply(map_age)
print(age_exp.isna().sum())

age_exp = age_exp.dropna(axis=0)
print(age_exp.isna().sum())
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(
    data=age_exp,
    x="Age",
    y="WorkExp"
)
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
query = """
SELECT Age, TimeSearching, Frustration FROM main
"""
age_time_frust = pd.read_sql(sql=query, con=conn)
age_time_frust.loc[:,"Age"] = age_time_frust.loc[:,"Age"].apply(map_age)
print(age_time_frust.isna().sum())

age_time_frust = age_time_frust.dropna(axis=0)
print(age_time_frust.isna().sum())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=age_time_frust,
    y="TimeSearching",
    x="Frustration",
    size="Age",       # scaled bubble size
    sizes=(100, 1000),       # min/max sizes
    hue="Age",               # optional: color by Age too
    palette="viridis",
    alpha=0.3
)

plt.title("Time Searching vs. Frustration (Bubble size = Age)")
plt.ylabel("Time Searching (units)")
plt.xlabel("Frustration (units)")
plt.xticks([])
plt.legend(title="Age", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
query = """
SELECT DatabaseWantToWorkWith FROM main
"""

databases_want = pd.read_sql(sql=query, con=conn)
top5 = (
    databases_want["DatabaseWantToWorkWith"]
      .dropna()
      .str.split(";")      # adjust delimiter if needed
      .explode()
      .str.strip()
      .value_counts()
      .head(5)
)

top5
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(x=top5.values, labels=top5.index, autopct='%1.1f%%', startangle=90,counterclock=False)

ax.set_title(label='Top 5 Databases Most Wanted to Learn', fontsize=14)
plt.tight_layout()
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
query = """
SELECT TimeSearching, TimeAnswering FROM main
WHERE Age = '25-34 years old';
"""

df_stacked = pd.read_sql(sql=query, con=conn)
print(df_stacked["TimeAnswering"].value_counts(dropna=False).sort_index())
print(df_stacked["TimeSearching"].value_counts(dropna=False).sort_index())
```

```python
# Load the data (replace this with your actual file path if needed)
df = pd.read_csv('survey-data.csv')

# Replace 'Prefer not to say' with NaN
df = df.replace('Prefer not to say', np.nan)

# Mapping for Age groups to their midpoints
mapping_age_group = {
    "25-34 years old": (25+34)/2, # 29.5
    "35-44 years old": (35+44)/2 # 39.5
}

# Map the Age column to age groups
df['AgeMapped'] = df['Age'].map(mapping_age_group)
df['AgeMapped'].unique()
```

```python
# Map the time categories to numeric values (minutes)
time_mapping = {
    'Less than 15 minutes a day': 7.5,  # midpoint of <15 minutes
    '15-30 minutes a day': 22.5,  # midpoint of 15-30 minutes
    '30-60 minutes a day': 45,  # midpoint of 30-60 minutes
    '60-120 minutes a day': 90,  # midpoint of 60-120 minutes
    'Over 120 minutes a day': 150  # assumed midpoint for >120 minutes
}

# Replace the time category columns with numeric values
df['TimeSearching'] = df['TimeSearching'].map(time_mapping)
df['TimeAnswering'] = df['TimeAnswering'].map(time_mapping)

# Drop rows where TimeSearching or TimeAnswering is NaN after mapping
df = df.dropna(subset=['TimeSearching', 'TimeAnswering'])
```

```python
# Replace NaN in AgeMapped with its mean value
df['AgeMapped'] = df['AgeMapped'].fillna(df['AgeMapped'].mean())

# Filter the data for age group 30 to 35
df_age_group = df[(df['AgeMapped'] >= 30.0) & (df['AgeMapped'] <= 35.0)]

# Calculate the median for TimeSearching and TimeAnswering
median_time_searching = df_age_group['TimeSearching'].median()
median_time_answering = df_age_group['TimeAnswering'].median()

# Prepare the data for the stacked bar chart 
data = {
    'TimeSearching': median_time_searching,
    'TimeAnswering': median_time_answering
}
```

```python
# Convert data into a DataFrame for plotting
df_median = pd.DataFrame([data])

# Plot the stacked bar chart
df_median.plot(kind='bar', stacked=True, figsize=(10, 6), color=['yellow', 'red'])

# Customize the plot
plt.title('Stacked Bar Chart of Median TimeSearching and TimeAnswering (Age 30-35)')
plt.xlabel('Age Group 30-35')
plt.ylabel('Time')
plt.legend(title="Time Type", labels=["TimeSearching", "TimeAnswering"])
plt.xticks([])  # Remove x-ticks as it's just one bar group
plt.grid(True, axis='y')

# Show the plot
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Example: load your dataframe
# df = pd.read_csv("your_file.csv")

# Count and sort categories
branch_counts = df['MainBranch'].value_counts(dropna=False)
branch_perc = (branch_counts / branch_counts.sum()) * 100

# Sort ascending for plotting so largest ends up on top after invert_yaxis
branch_counts = branch_counts.sort_values(ascending=True)
branch_perc = branch_perc[branch_counts.index]

# Create distinct colours for each bar
colors = plt.cm.tab20(np.linspace(0, 1, len(branch_counts)))  # 'tab20' is nice and varied

fig, ax = plt.subplots(figsize=(16, 5))

# Plot horizontal bars with unique colours
bars = ax.barh(
    range(len(branch_counts)),
    branch_counts.values,
    color=colors,
    edgecolor='black'
)

# Remove y-axis ticks/labels
ax.set_yticks([])

# Add legend mapping colours to categories
ax.legend(
    bars,
    branch_counts.index,
    title="MainBranch",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

# Add percentage labels to the right of each bar
for i, (count, pct) in enumerate(zip(branch_counts.values, branch_perc.values)):
    ax.text(
        count + branch_counts.max() * 0.01,  # offset
        i,
        f"{pct:.1f}%",
        va='center'
    )

# Labels and title
ax.set_xlabel("Respondent Count", fontsize=12)
ax.set_title("MainBranch Distribution", fontsize=14)

# Put largest bar on top
ax.invert_yaxis()

plt.tight_layout()
plt.show()
```
